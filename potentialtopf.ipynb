```python
%matplotlib inline
from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

h_bar = 1.0546e-34
me = 9.109e-31
q=1.602e-19

def func_Eges(n,a):
    a=a*10e-11
    Eges=((h_bar**2*np.pi**2)/(2*me*a**2)*n**2)/q
    return Eges

def func_psi(n,a):
    psi=(np.sqrt(2/a)*np.sin((n*np.pi)/a*x))**2
    return psi
```

```python
a = float(input("Moleküllänge eingeben: "))
n = float(input("Quantenzahl eingeben: "))
x = np.arange(0,a,0.1)
print("Eges = ", func_Eges(n,a))    
plt.plot(x,func_psi(n,a))
```
